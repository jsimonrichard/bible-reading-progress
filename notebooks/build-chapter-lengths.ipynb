```python
from pysword.modules import SwordModules
# Load module in zip
# NB: the zip content is only available as long as the SwordModules object exists
modules = SwordModules(u'/home/jsimonrichard/Downloads/KJV.zip')
# In this case the module found is:
# {'KJV': {'description': 'KingJamesVersion(1769)withStrongsNumbersandMorphology', 'encoding': 'UTF-8', ...}}
found_modules = modules.parse_modules()
bible = modules.get_bible_from_module(u'KJV')
```

```python
book_structure = bible.get_structure().get_books()
structure_serialized = {}
for group in book_structure:
    structure_serialized[group] = {}
    for book in book_structure[group]:
        structure_serialized[group][book.name] = book.chapter_lengths
structure_serialized
```

```python
import json

with open('../bible_structure.json', 'w') as f:
    json.dump(structure_serialized, f)
```
